```python
import math as m

#def h(x):
#    return 1/(1+m.e(-x))
def sigmoid(z):
    return 1 / (1 + np.exp(-z))


def hx(w,X):
    z = np.array(np.multiply(w[0],np.array(X[:,0])) + np.multiply(w[1],np.array(X[:,1])) + np.multiply(w[2],np.array(X[:,2])))
    return sigmoid(z)

def grad(theta,X,Y):
    y_pred = hx(theta,X)
    g = [0]*len(theta)
    g=np.matrix(g).T
    #print(g.shape)
    for i in range(len(theta)):
        g[i] = -1 * sum(np.multiply(np.multiply(Y.T,(1-y_pred)),X[:,i]) - np.multiply(np.multiply(1-Y.T,y_pred),X[:,i]))
    return g

def cost(theta,a,b):
    ans=0
    for i in range(len(a)):
        k=a[i,:]*theta
        k=m.log(h(k))
        if y[i]==1 :
            ans=ans+k
        else :
            ans=ans+1-k
    return ans


a=0.001
theta=np.random.rand(3,1)
oldtheta=theta
x1=np.matrix(x1)
y1=np.matrix(y1)
#print(y1)
for i in range (1000):
    oldtheta=theta
    theta=theta-a*grad(theta,x1,y1)

print(theta)
```
